```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('TSP_results/results.csv', index_col=0, skiprows=[10,11,16,19,20,21,22,23,24,26,27,28,30,40,41,42])
```

```python
df
```

```python
methonds_name = ['NN_cost','NN2Opt_cost', 'NN2OptDLB_cost', 'NN3Opt_cost','NN3OptDLB_cost','repNN_cost','repNN2Opt_cost','repNN2OptDLB_cost','repNN3Opt_cost','repNN3OptDLB_cost']
df_2 = df.loc[:,methonds_name]
```

```python
df_2
```

```python
costs = df_2.to_numpy()
diff_cost = np.zeros((len(costs), len(costs[0])))
```

```python
costs
```

```python
for i, row in enumerate(costs):
    for j in range(len(row)):
        diff_cost[i][j] = (costs[i][j] - df['opt_result'].iloc[i]) / df['opt_result'].iloc[i]
```

```python
sns.heatmap(diff_cost, xticklabels=methonds_name, yticklabels=df['name'], cmap='viridis_r')
plt.xticks(rotation=45, ha='right', rotation_mode='anchor')
plt.title('Deviation from optimal solution')
plt.show()
```
